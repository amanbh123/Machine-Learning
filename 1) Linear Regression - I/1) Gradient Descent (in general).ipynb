```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2
print(X, Y)
```

```python
plt.style.use("seaborn")
```

### Given a function f(x), we want to find the value of x that minimises f(x)

```python
plt.plot(X, Y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gradient Descent")
plt.show()
```

```python
x = 0
lr = 0.1
vals = []
plt.plot(X, Y, color = 'orange')

# 50 steps in downward direction
for i in range(50):
    grad = 2*(x-5)
    x = x - lr * grad
    y = (x-5)**2
    vals.append(y)
    plt.scatter(x, y, color = 'black')

plt.show()    
```
